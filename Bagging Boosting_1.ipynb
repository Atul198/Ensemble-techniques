**Ensemble Techniques**

- Bagging - Building multiple models (typically of the same type) from different subsamples of the training dataset.
- Boosting - Building multiple models (typically of the same type) each of which learns to fix the prediction errors of a prior model in the chain.
- Voting - Building multiple models (typically of differing types) and simple statistics (like calculating the mean) are used to combine predictions.

## Outliers Analysis

```python
continious_var = df.select_dtypes(include = ['int64', 'float64'])

plt.figure(figsize = (15, 8))
df.boxplot(column = list(continious_var), rot=65)
plt.show()
```

```python
def outlier_lims(col):
    q3, q1 = np.percentile(col, [75, 25])
    iqr = q3 - q1
    upper_lim = q3 + 1.5 * iqr
    lower_lim = q1 - 1.5 * iqr
    return upper_lim, lower_lim
```

```python
outliers_list = []
for col in continious_var:
    print('-------------------------------------------')
    print("Column:", col)
    
    UL,LL = outlier_lims(df[col])
    print("Upper Limit =", UL)
    print("Lower Limit =", LL)
    
    total_outliers = len(df.loc[df[col]<LL,col]) + len(df.loc[df[col]>UL,col])
    percent = (total_outliers / len(df.index) )*100

    if percent < 10 and percent!=0:
        outliers_list.append(col)
    
    print("Percentage of Outliers=", percent)
    print("-------------------------------------------------- \n")

print(outliers_list)
```

['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'BedroomAbvGr', 'KitchenAbvGr', 'Fireplaces', 'GarageCars', 'WoodDeckSF', 'OpenPorchSF', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal']

```python
for col in outliers_list:
    UL, LL = outlier_lims(df[col])
    df.loc[df[col] > UL, col ] = np.nan
    df.loc[df[col] < LL, col] = np.nan

df.dropna(subset=outliers_list, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# Reducing Skewness from dataset
df[outliers_list] = np.log1p(df[outliers_list])
```

## Label Encoding

## Standerdization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

continious_var = data.select_dtypes(include=['int64', 'float64']).columns


data[continious_var] = scaler.fit_transform(data[continious_var])

data.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X = data.drop(['SalePrice', 'Id'], axis=1)
y = data['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

## Decision Tree Regression, Random Forest Regreession, Bagging Regressor, Ada Boosting , Voting

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor, VotingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

# Define a function to evaluate the model
def evaluate_model(y_test, predictions, model_name):
    mae = mean_absolute_error(y_test, predictions)
    mse = root_mean_squared_error(y_test, predictions)
    rmse = root_mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    
    print(f"\n{model_name} Performance Metrics:")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"R² Score: {r2:.4f}")
    
    # Scatter plot
    plt.figure(figsize=(6, 6))
    plt.scatter(y_test, predictions, color='blue', label='Predicted vs Actual')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Ideal Line')
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"Actual vs Predicted ({model_name})")
    plt.legend()
    plt.show()

# Bagging: Decision Trees and Random Forest
dt = DecisionTreeRegressor()
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Boosting: AdaBoost and Gradient Boosting
ada = AdaBoostRegressor(estimator=DecisionTreeRegressor(), n_estimators=50, random_state=42)
gb = GradientBoostingRegressor(n_estimators=100, random_state=42)
xgb = XGBRegressor(n_estimators=100, random_state=42)

# Voting: Combining multiple models
voting_regressor = VotingRegressor(estimators=[
    ('dt', dt),
    ('rf', rf),
    ('gb', gb),
    ('xgb', xgb)
])

# List of models
models = [
    ('Decision Tree', dt),
    ('Random Forest', rf),
    ('AdaBoost', ada),
    ('Gradient Boosting', gb),
    ('XGBoost', xgb),
    ('Voting Regressor', voting_regressor)
]

# Fit and evaluate each model
for model_name, model in models:
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    evaluate_model(y_test, predictions, model_name)
```
